# Gaussian Distribution 

The Normal distribution is also called the gaussian distribution. This distribution 
is most important distributionto talk about. We can use the random.normal() method to get a Normal Data Distribution. 

For example: IQ scores and Heartbeats arrays

It has three paramaters:
- **loc** - (Mean) where the peak of the bell exists 
- **scale** - (Standard Deviation) How flat the graph distribution should be 
- **size** - The shape of the returned array, 

```python
import numpy as np
from numpy import random
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 

# random normal distribution of size 2*3 with mean at 1 and standard deviation of 2. 
x = random.normal(size=(5, 40))
print(x)
```

```python
df = pd.DataFrame(x)
```

```python
plt.figure(figsize=(12, 7))
plt.style.use('default')

sns.kdeplot(data=x[0], label="Gauss 1", color='red')
sns.kdeplot(data=x[1], label='Gauss 2', color='blue')
sns.kdeplot(data=x[2], label='Gauss 3', color='purple', multiple='stack')
sns.kdeplot(data=x[3], label='Gauss 4', color='green')
sns.kdeplot(data=x[4], label='Gauss 5', color='orange', multiple='stack')

plt.title('Distribution', fontsize=18, fontweight='bold')
plt.xlabel("Gaussian Distribution", fontsize=16, fontweight='semibold')
plt.ylabel('Density', fontsize=16, fontweight='semibold')

plt.legend(shadow=True, fancybox=True, loc="upper right", title="gaussian")
plt.grid(True, alpha=0.3, linestyle="--")
```
