```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
import pandas as pd
```

```python
# plotting a distplot 
sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
# Plotting a Distplot without the Histogram 
sns.displot([0, 1, 2, 3, 4, 5], kind='kde')
plt.show()
```

```python
# Original small dataset values
heights = [150,160,165,170,175,180,155,168,172,178]
weights = [50,55,62,65,70,75,52,63,68,74]
ages    = [20,22,25,28,30,32,21,27,29,31]
scores  = [70,75,80,85,88,90,72,82,86,89]

# Create 100 rows by sampling with some noise
np.random.seed(42)  # reproducibility
rows = 500

data = {
    "Height": np.random.choice(heights, rows) + np.random.randint(1, 100, rows),
    "Weight": np.random.choice(weights, rows) + np.random.randint(1, 50, rows),
    "Age":    np.random.choice(ages, rows) + np.random.randint(1, 25, rows),
    "Score":  np.random.choice(scores, rows) + np.random.randint(1, 15, rows),
}

dataset = pd.DataFrame(data)
print(dataset)
```

## Scatter Plot

```python
plt.figure(figsize=(12, 7))
plt.style.use('default')

colors = ['red', 'green', 'yellow', 'blue']
labels = ['cluster 1', 'cluster 2', 'cluster 3', 'cluster 4']

plt.scatter(data['Height'], data['Weight'], color=colors[0], label=labels[0], s=50, edgecolor='white', linewidth=0.8, alpha=0.7)
plt.scatter(data['Weight'], data['Score'], color=colors[1], label=labels[1], s=50, edgecolor='white', linewidth=0.8, alpha=0.7)
plt.scatter(data['Score'], data['Height'], label=labels[2], color=colors[2], s=50, edgecolor='white', linewidth=0.8, alpha=0.7)
plt.scatter(data['Age'], data['Weight'], label=labels[3], color=colors[3], s=50, edgecolor='white', linewidth=0.8, alpha=0.5)

plt.title("Random Clusterization", fontsize=18, fontweight='bold', pad=20)
plt.xlabel("X_component", fontsize=16, fontweight='semibold')
plt.ylabel("y_component", fontweight="semibold", fontsize=16)

plt.legend(shadow=True, title="Clusters", loc="upper right", fancybox=True)
plt.grid(True, alpha=0.3, linestyle='--')
plt.show()
```

## Histplot 

```python
# handaling distributions word counts 
plt.figure(figsize=(10, 7))
sns.histplot(data['Height'], kde=True, label='Height', color='blue')
sns.histplot(data['Weight'], kde=True, label='Weight', color='red')

plt.title('Distributions of Heights VS Weights', fontsize=12, fontweight='bold')
plt.xlabel('Comparision of Weights and Heights', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=14, fontweight='bold')
plt.legend(shadow=True, title='summary', loc='upper right')
plt.show()
```

# Correlation

```python
corr_matrix=dataset[['Height', 'Weight', 'Score', 'Age']].corr()

sns.heatmap(corr_matrix, annot=True, fmt='0.2f', linewidth=0.5, cmap='coolwarm')

plt.xticks(rotation=45)
plt.show()
```

```python
corr_matrix = dataset[['Height', 'Weight', "Score", 'Age']].corr()

sns.set(font_scale=1.2)
sns.heatmap(corr_matrix, annot=True, fmt='0.2f', linewidths=0.5, cmap='coolwarm')

plt.title('Correlation Matrix Heatmap', fontsize=14, fontweight='bold')

plt.xticks(rotation=45)
plt.show()
```
