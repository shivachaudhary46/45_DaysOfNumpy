# Binomial Distribution 

Binomial distribution is a Discrete distribution. 

It discribes the outcome of binary scenarios, for eg: toss of a coin because it can be 
either heads or tails

It has three parameters:

- n = number of trials

- p = probability of occurence of each trial (for eg: toss of coin 0.5 each)

- size = shape of array need to be returned

for example: The distribution is defined at seperate set of events. E.g , a coin toss's result is discrete as it can be only head or tails. whereas the height of people is continuous as it can be 170, 170.1, 170.4, 170.11 and so on. 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# give a 10 trials for coin toss and generate 10 data points.

from numpy import random

x=random.binomial(n=100, p=0.5, size=100)
print(x)
```

# Visualization of Binomial Distribution

```python
plt.style.use('default')
plt.Figure(figsize=(12, 7))

sns.histplot(x, kde=True, color="blue", multiple='stack')

plt.xlabel("distribution", fontsize=14, fontweight="semibold")
plt.ylabel("density", fontsize=14,fontweight='semibold' )
plt.title("Visualization of Binomial distribution", fontsize=16, fontweight='bold')

plt.show()
```

```python
normal = random.normal(loc=5, scale=3, size=(2, 5))
bin = random.binomial(n=10, p=0.5, size=10)
```

```python
print(normal)
print(bin)
```

```python
plt.Figure(figsize=(12, 7))
plt.style.use("default")

sns.kdeplot(data=normal[0], label="gauss 1", color='red')
sns.kdeplot(data=normal[1], label="gauss 2", color="green")
sns.kdeplot(data=bin, label="Binomial dist", color="blue")

plt.title("Gaussian VS Binomial", fontsize=18, fontweight="bold")
plt.xlabel("distributions", fontsize=14, fontweight="semibold")
plt.ylabel("density", fontsize=14,fontweight='semibold' )

plt.legend(shadow=True, title="Types", fancybox=True, loc="upper left")
plt.grid(True, alpha=0.5, linestyle="--")
plt.show()
```
