## Uniform distribution 
It used to describe the probability where every event has equal chances of happening or occuring. 

E.g. Generation of random numbers, if we want to generate the random array size of the (1, 3) range from -3 to 3 then, the probability of occuring the number in the array ranging will be equal. 

It has three parameters: 

- low - lower bound - default 0.0.
- high - upper bound - default 1.0 
- size - The shape of the returned array 

```python
from numpy import random 

x = random.uniform(size=(2, 3))

print(x)
```

## Visualization of Uniform Distribution

```python
from numpy import random 
import matplotlib.pyplot as plt
import seaborn as sns 

sns.kdeplot(random.uniform(size=1000), color="blue", multiple="stack", label="unifrom dist")
sns.kdeplot(random.poisson(size=1000), color="yellow", label='poisson dist', multiple='stack')

plt.title("Poisson VS Uniform distribution", fontsize=16, fontweight="bold")
plt.xlabel("distributions", fontsize=14, fontweight="semibold")
plt.ylabel("density", fontsize=14, fontweight="semibold")

plt.legend(shadow=True, loc="upper right", title="distribution comparision")
plt.grid(True, alpha=0.5, linestyle='--')
plt.show()
```
