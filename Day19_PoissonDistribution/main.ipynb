# Possion Distribution 

It estimate how many times an event can happen in a specified time. e.g. if someone eats twice a day what is the probability he will eat thrice ?

**NOTE: Poission Distribution is Discrete Distribution**

- lam - rate or known number of occurences e.g. 2 for above problem.
- size - the shape of returned array 

```python
from numpy import random 
z = random.poisson(lam=2, size=5)
print(z)
```

# Visualization of the Poisson Distribution

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(random.poisson(lam=2, size=1000), kde=True, color="blue")

plt.show()
```

## Normal and Poisson Distribution 

```python
gauss = random.normal(loc=50, scale=7, size=1000)
poiss = random.poisson(lam=50, size=1000)

plt.Figure(figsize=(12, 7))
plt.style.use("default")

sns.kdeplot(data=gauss, label="Normal Distri", color="blue", multiple='stack')
sns.kdeplot(data=poiss, label="Poiss Distri", color="red", multiple="stack")

plt.title("Normal VS Poisson", fontsize=18, fontweight="bold")
plt.xlabel("distributions", fontsize=14, fontweight="semibold")
plt.ylabel("density", fontsize=14,fontweight='semibold' )

plt.legend(shadow=True, loc="upper right", title="Types")
plt.grid(True, alpha=0.5, linestyle="--")
plt.show()
```
